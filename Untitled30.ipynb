```python
# @title ðŸ–¼ï¸ Image Upload & Enhancement Tool

# 1. Install required packages
!pip install -q torch==2.0.1+cpu torchvision==0.15.2+cpu -f https://download.pytorch.org/whl/torch_stable.html
!pip install -q basicsr==1.4.2 realesrgan==0.3.0 opencv-python==4.8.1.78

# 2. Import libraries
from google.colab import files
import cv2
import numpy as np
from IPython.display import display, Image
from matplotlib import pyplot as plt
from basicsr.archs.rrdbnet_arch import RRDBNet
from realesrgan import RealESRGANer
import torch
import os

# 3. Initialize the upscaler
print("âš™ï¸ Initializing upscaler...")
model = RRDBNet(num_in_ch=3, num_out_ch=3, num_feat=64, num_block=23, num_grow_ch=32, scale=4)
upsampler = RealESRGANer(
    scale=4,
    model_path='/content/Real-ESRGAN/weights/RealESRGAN_x4plus.pth',
    model=model,
    tile=0,
    tile_pad=10,
    pre_pad=0,
    half=False,
    device=torch.device('cpu')
)

# 4. File upload function
def upload_and_enhance():
    print("\nðŸ“¤ Step 1: Upload your image")
    uploaded = files.upload()
    if not uploaded:
        print("âŒ No file uploaded!")
        return

    filename = list(uploaded.keys())[0]
    input_path = f'/content/{filename}'

    # Read the image
    img = cv2.imread(input_path)
    if img is None:
        print("âŒ Failed to read the image!")
        return

    print(f"\nðŸ–¼ï¸ Original Image ({img.shape[1]}x{img.shape[0]})")
    display(Image(filename=input_path, width=400))

    # Enhance the image
    print("\nâš¡ Enhancing image...")
    output, _ = upsampler.enhance(img, outscale=4)
    output_path = f'/content/enhanced_{filename}'
    cv2.imwrite(output_path, output)

    # Show results
    print(f"\nâœ¨ Enhanced Image ({output.shape[1]}x{output.shape[0]})")
    plt.figure(figsize=(20, 10))
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.subplot(1, 2, 2)
    plt.title("Enhanced 4x")
    plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
    plt.show()

    # Download link
    print("\nâœ… Enhancement complete!")
    print(f"ðŸ“¥ Download enhanced image:")
    from IPython.display import FileLink
    display(FileLink(output_path))

# 5. Run the uploader
upload_and_enhance()
```

```python
# @title ðŸ”‡ Working Image Denoising Tool
from google.colab import files
import cv2
import numpy as np
from IPython.display import display, Image, FileLink
from basicsr.archs.rrdbnet_arch import RRDBNet
from realesrgan import RealESRGANer
import torch

def denoise_image():
    print("ðŸ“¤ Upload a noisy image:")
    uploaded = files.upload()
    if not uploaded:
        print("âŒ No file uploaded!")
        return

    # Load image
    filename = next(iter(uploaded))
    img = cv2.imdecode(np.frombuffer(uploaded[filename], np.uint8), cv2.IMREAD_COLOR)

    # Initialize with correct parameters
    model = RRDBNet(num_in_ch=3, num_out_ch=3, num_feat=64, num_block=23, num_grow_ch=32, scale=4)
    denoiser = RealESRGANer(
        scale=4,
        model_path='/content/Real-ESRGAN/weights/RealESRGAN_x4plus.pth',
        model=model,
        tile=400,  # Helps with memory
        tile_pad=10,
        pre_pad=0,
        half=False,
        device=torch.device('cpu')
    )

    print("ðŸ”§ Denoising...")
    # Use enhance with outscale=1 for denoising (no upscaling)
    denoised, _ = denoiser.enhance(img, outscale=1)

    # Save and show
    cv2.imwrite('denoised.jpg', denoised)
    print("\nBefore vs After:")
    display(Image(filename=filename, width=300))
    display(Image('denoised.jpg', width=300))
    display(FileLink('denoised.jpg'))

denoise_image()
```

```python
# @title ðŸŽ¨ Single-Image Inpainting (Paint Mask Directly)
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import ipywidgets as widgets

# 1. Upload Image
print("ðŸ“¤ Upload your image:")
uploaded = files.upload()
if not uploaded:
    print("âŒ No file uploaded!")
else:
    # Load image
    img = cv2.imdecode(np.frombuffer(next(iter(uploaded.values())), np.uint8), cv2.IMREAD_COLOR)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    height, width = img.shape[:2]

    # 2. Create Drawing Interface
    plt.figure(figsize=(10,10))
    plt.imshow(img)
    plt.title("Draw on the image (close window when done)")
    plt.draw()

    # Enable drawing
    def on_click(event):
        if event.inaxes:
            radius = 15
            circle = plt.Circle((event.xdata, event.ydata), radius, color='black')
            ax = event.inaxes
            ax.add_patch(circle)
            plt.draw()

    plt.gcf().canvas.mpl_connect('button_press_event', on_click)
    plt.show()

    # 3. After drawing, get the mask
    print("ðŸ”„ Processing your drawing...")
    mask = np.zeros((height, width), dtype=np.uint8)
    for patch in plt.gca().patches:
        x, y = patch.center
        radius = patch.radius
        cv2.circle(mask, (int(x), int(y)), int(radius), 255, -1)

    # 4. Inpainting
    result = cv2.inpaint(
        cv2.cvtColor(img, cv2.COLOR_RGB2BGR),
        mask,
        inpaintRadius=3,
        flags=cv2.INPAINT_NS
    )

    # 5. Show Results
    clear_output()
    plt.figure(figsize=(15,5))
    plt.subplot(1,3,1); plt.imshow(img); plt.title("Original")
    plt.subplot(1,3,2); plt.imshow(mask, cmap='gray'); plt.title("Your Mask")
    plt.subplot(1,3,3); plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
    plt.title("Result")
    plt.show()

    # Save results
    cv2.imwrite('result.jpg', result)
    print("âœ… Done! Right-click images below to save:")
    display(Image('result.jpg', width=600))
```

```python
# @title ðŸ–¼ï¸ All-in-One Image Toolkit (Fixed Version)

# 1. SETUP
!pip install -q ipywidgets opencv-python matplotlib
from google.colab import files
import cv2, numpy as np
from IPython.display import display, clear_output
import ipywidgets as widgets
from PIL import Image
import matplotlib.pyplot as plt
import io

# 2. CORE FUNCTIONS
def enhance(img):
    """ Basic sharpening """
    kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
    return cv2.filter2D(img, -1, kernel)

def colorize(img):
    """ Simple color tinting """
    img = img.astype(np.float32)
    img[:,:,0] *= 1.3  # Red
    img[:,:,1] *= 0.9  # Green
    img[:,:,2] *= 0.8  # Blue
    return np.clip(img, 0, 255).astype(np.uint8)

def denoise(img):
    """ OpenCV denoising """
    return cv2.fastNlMeansDenoisingColored(img, None, 10, 10, 7, 21)

def inpaint(img, mask):
    """ Basic inpainting """
    return cv2.inpaint(img, mask, 3, cv2.INPAINT_TELEA)

# 3. INTERACTIVE UI
upload_btn = widgets.FileUpload(description="ðŸ“¤ Upload Image", accept="image/*")
process_dropdown = widgets.Dropdown(
    options=['Enhance', 'Colorize', 'Denoise', 'Inpaint'],
    description='ðŸ› ï¸ Action:',
    style={'description_width': '50px'}
)
run_btn = widgets.Button(description="âœ¨ Process", button_style='success')
output = widgets.Output()
mask = None

# Create drawing interface
def create_mask(img):
    global mask
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,6))
    ax1.imshow(img)
    ax1.set_title("Original")

    # Initialize empty mask
    mask = np.zeros(img.shape[:2], dtype=np.uint8)
    ax2.imshow(mask, cmap='gray')
    ax2.set_title("Draw Mask (Black=Remove)")

    def on_click(event):
        if event.inaxes == ax2:
            x, y = int(event.xdata), int(event.ydata)
            cv2.circle(mask, (x,y), 15, 255, -1)
            ax2.imshow(mask, cmap='gray')
            fig.canvas.draw()

    fig.canvas.mpl_connect('button_press_event', on_click)
    plt.show()

def on_run_click(b):
    global mask
    with output:
        clear_output()
        if not upload_btn.value:
            print("âŒ Please upload an image first")
            return

        # Load image (FIXED file handling)
        img_data = upload_btn.data[0]
        img = cv2.imdecode(np.frombuffer(img_data, np.uint8), cv2.IMREAD_COLOR)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Process
        choice = process_dropdown.value
        if choice == 'Enhance':
            result = enhance(img)
        elif choice == 'Colorize':
            result = colorize(img)
        elif choice == 'Denoise':
            result = denoise(img)
        elif choice == 'Inpaint':
            create_mask(img)
            return

        # Display results
        if choice != 'Inpaint':
            plt.figure(figsize=(12,6))
            plt.subplot(121); plt.imshow(img); plt.title("Original")
            plt.subplot(122); plt.imshow(result); plt.title("Result")
            plt.show()

        # Handle inpainting after mask creation
        if mask is not None:
            result = inpaint(img, mask)
            plt.figure(figsize=(12,6))
            plt.subplot(131); plt.imshow(img); plt.title("Original")
            plt.subplot(132); plt.imshow(mask, cmap='gray'); plt.title("Mask")
            plt.subplot(133); plt.imshow(result); plt.title("Inpainted")
            plt.show()
            mask = None

run_btn.on_click(on_run_click)
display(widgets.VBox([
    widgets.HBox([upload_btn, process_dropdown]),
    run_btn,
    output
]))
```

```python
# @title ðŸŒ 100% Working Web Server (Run This First)
# 1. Install EXACT versions needed
!pip install -q flask==2.3.2 flask-ngrok==0.1.0 flask-cors==4.0.0 pillow opencv-python

# 2. Create server
from flask import Flask, request, send_file
from flask_ngrok import run_with_ngrok
from flask_cors import CORS
import cv2, numpy as np
import io
from PIL import Image

app = Flask(__name__)
CORS(app)
run_with_ngrok(app)  # This creates the public URL

# Image processing functions
def enhance(img):
    kernel = np.array([[0,-1,0], [-1,5,-1], [0,-1,0]])
    return cv2.filter2D(img, -1, kernel)

@app.route('/process', methods=['POST'])
def process():
    file = request.files['image']
    img = cv2.imdecode(np.frombuffer(file.read(), np.uint8), cv2.IMREAD_COLOR)
    result = enhance(img)  # Demo with just enhancement
    _, img_encoded = cv2.imencode('.jpg', result)
    return send_file(io.BytesIO(img_encoded), mimetype='image/jpeg')

print("âœ… Server ready! Access via the URL that appears below â†“")
app.run()
```
